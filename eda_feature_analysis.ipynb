```python
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
pd.set_option('display.max_columns', 50)
```

```python
df = pd.DataFrame.from_csv('train.csv')
df
```

```python
# Creates a dataframe for each hazard index and the sum of the first two column means
y = []
for rating in np.sort(pd.unique(df.Hazard)):
    y.append(df[df['Hazard'] == rating]['T1_V1'].mean() + df[df['Hazard'] == rating]['T1_V1'].mean())

x = np.sort(pd.unique(df.Hazard))

df1 = pd.DataFrame({'x':x, 'y': y})
```

```python
#Plots the mean 2 column sum for each hazard rating
plot = sns.lmplot(x='x',y ='y', data=df1)
plot.set_axis_labels("Hazard Rating", "Mean comined first columns")
plot.fig.suptitle('Mean 2 col sum per Hazard Rating')
```
